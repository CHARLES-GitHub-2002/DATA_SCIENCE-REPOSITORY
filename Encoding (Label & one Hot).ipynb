# Encoding

```python
#importing the essential Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
#reading the files 
data = pd.read_excel(r"C:\Users\CHARLES\Desktop\Book1.xlsx")
```

```python
#looking at first five rows of the dataset
data.head()
```

```python
#descriptive statistic summary of the dataset
data.describe().T
```

```python
#cheaking for null values in the dataset
data.isnull().sum()
```

```python
#using the mean strategy to impute the missing value using simpleimputer class
imputer=SimpleImputer(missing_values=np.NaN)
imp=imputer.fit(data.iloc[:,1:3])
data.iloc[:,1:3]=imp.transform(data.iloc[:,1:3])
```

```python
data.head()
```

```python
## Label Encoding
from sklearn import preprocessing
encoder=preprocessing.LabelEncoder()
data['Purchased']=encoder.fit_transform(data['Purchased'])
```

```python
data.head()
```

```python
#one hot Encoding 
county=data[['County']].values
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
county=encoder.fit_transform(county).toarray()
county=pd.DataFrame(county)
county_columns=['france','germany','india','spain']
county.columns=county_columns
final_df=county.join(data)
final_df.drop(columns=('County'),axis=1,inplace=True)
```
